```python
import pandas as pd
```

```python
df1 = pd.read_csv('acc_child.csv',encoding='cp949')
```

```python
df1.iloc[1,:] == '소계' # 소계로 되어 있는 부분 찾기
```

```python
df2 = df1.drop(df1.loc[df1['행정구역별(2)'] == '소계'].index) # 소계로 되어 있는 컬럼 제거
```

```python
df3 = df2.drop(df2.index[0])
```

```python
df4 = df3.rename(columns={'행정구역별(1)' : "지역", "행정구역별(2)" : "구/군", "2021":"2021_사고건수","2021.1":"2021_사망자수","2021.2":"2021_부상자수", "2022":"2022_사고건수","2022.1":"2022_사망자수","2022.2":"2022_부상자수"}) # 컬럼명 변경
```

```python
df4
```

```python
df5 = df4.set_index(['지역','구/군'])
df5 # 지역과 구/군으로 멀티 인덱스
```

```python
for column in df5.columns:
    df5[column] = df5[column].replace('-','0')
df5 # 결측값 0으로 치환
```

```python
df5.info()
```

```python
df5
```

```python
df6 = df5.reset_index()
```

```python
df5
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
for column in df5.columns:
    df5[column] = pd.to_numeric(df5[column],errors='coerce')
df5 # 전체 데이터 숫자로 변경
```

```python
df5.info()
df6 = df5.reset_index()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="지역",y="2021_사고건수",data=df6)
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
pivot_table = df6.pivot_table(columns='지역', values='2021_사고건수', aggfunc='sum') # 지역별 2021년 사고 건수의 분포와, 사고건수 합 테이블
pivot_table
```

```python
pivot_table = df6.pivot_table(columns='지역', values='2021_사망자수', aggfunc='sum')
pivot_table # 지역별 사망자수 테이블
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="지역",y="2021_부상자수",data=df6)
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
pivot_table = df6.pivot_table(columns='지역', values='2021_부상자수', aggfunc='sum') # 지역별 2021년 부상자수 분포와 부상자수 합 테이블
pivot_table 
```
